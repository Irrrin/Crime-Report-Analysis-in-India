# Data Visualization

1. Scatter Plot Matrix

```python
sns.pairplot(df)
plt.show()
```

2. Bar Chart


```python
import matplotlib.pyplot as plt
 
plt.figure(figsize=(10,6))
plt.bar(df['State_UT'], df['Rate_Of_Total_Crime _Against_Women_2021'])
plt.xlabel('State/UT')
plt.ylabel('Rate of Total Crime Against Women (2021)')
plt.title('Total Crime Against Women by State')
plt.xticks(rotation=90)
plt.show()
```

And above Bar chart shows the total number of crime happens in each state.
* There is high number of crime against women happening in ASSAM and next state DELHI.
* There is lowest number of Crime happening in NAGALAND.
* Our State KERALA showing below 80.

```python
top_5_states = df.nlargest(5, 'Rate_Of_Total_Crime _Against_Women_2021')
plt.figure(figsize=(10,6))
plt.bar(top_5_states['Category'], top_5_states['Rate_Of_Total_Crime _Against_Women_2021'])
plt.xlabel('State/UT')
plt.ylabel('Rate of Total Crime Against Women (2021)')
plt.title('Top 5 states with Highest Crime Rate')
plt.xticks(rotation=90)
plt.show()
```

3. Scatter Plot

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plt.scatter(df['Mid_Year_Projected_Population_In_Lakhs_2021'],df['Rate_Of_Total_Crime _Against_Women_2021'], c='red')
plt.xlabel('Mid-Year project population (2021)')
plt.ylabel('Rate of Total Crime Against Women(2021)')
plt.title('Population vs Crime Rate')
plt.show()
```

Above there s the data points are scattered with a general downward trend as the population ncreases. The majority of the data points are concentrated btween a population of 0 and 2000 with corresponding crime rates ranging from approximately 20 to 100. There are two outliers on the right side of the graph, representing ppulations of around 5500 and 6000 with crime rates of approximately 50 and 60.

4. Heatmap Correlation

```python
corr_matrix = df[['Year_2019', 'Year_2020','Year_2021','Mid_Year_Projected_Population_In_Lakhs_2021','Rate_Of_Total_Crime _Against_Women_2021']].corr()
plt.figure(figsize=(10,8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('correlation Between Columns')
plt.show()
```

"The Heatmap illustrates a strong positive correlation among the columns,'Year_2019','Year_2020', and 'Year_2021','Mid_Year_Projected_Population_in_Lakhs_2021','Rate_Of_Crime_Against_Women' with correlation cofficient close to 1. This suggests that the data across these years exhibities a highly consistent pattern. The Correlation between these years and 'Mid_Year_Projected_Population_In_Lakhs_2021' is positive but slightly weaker, ranging from 0.86 to 0.88 for the years and around 0.5 for the direct correlation with 'Mid_Year_Projected_Population_In_Lakhs_2021', The weakest correlation is observed between 'Mid_Year_Projected_Population_In_Lakhs_2021' and other columns, with values around 0.5, indicating a moderate positive relationship".

5. Line PLOT

```python
plt.figure(figsize=(10,6))
for index, row in df.iterrows():
    plt.plot(['2019','2020','2021'],[row['Year_2019'],row['Year_2020'],row['Year_2021']],label=row['State_UT'])
    plt.xlabel('Year')
    plt.ylabel('Crime Rate')
    plt.title('Crime Trend Over Years')
    plt.legend()
    plt.show()
```

6. Box Plot

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='Category', y='Rate_Of_Total_Crime _Against_Women_2021', data=df)
plt.xlabel('State/UT')
plt.ylabel('Rate of Total Crime Against Women(2021)')
plt.title('Distribution of crime rate by state')
plt.xticks(rotation=90)
plt.show()
```

7. Violin Plot

```python
plt.figure(figsize=(10,6))
sns.violinplot(x='Category', y='Chargesheeting_Rate_2021', data=df)
plt.xlabel('State/UT')
plt.ylabel('Chargesheeting Rate (2021)')
plt.title('Distribution of Chargesheeting rate by state')
plt.xticks(rotation=90)
plt.show()
```

Violin plot, which is a type of Statistical graph used to visualize the distribution of dataset. The plot is comparing the "Chargesheeting Rate (2021)" for two categories: States and Union Territories.
* The distribution of chargesheeting rates for States appears to be slightly skewed to the right with a median around 70-80.
* The distribution of chargesheeting rates for Union Territories is more symmetric with a median around 80-90.
